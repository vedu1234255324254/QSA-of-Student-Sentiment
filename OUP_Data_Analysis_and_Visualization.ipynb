```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

def create_oup_dataset():
    """Create the OUP India dataset with 120 excerpts including source and without location"""

    # Sample excerpts for different categories and sentiments
    excerpts_data = [
        # Technical Issues (Negative)
        ["OUP_001", "Oxford Educate app keeps crashing when I try to download engineering textbooks for offline study", "Reddit", "Technical Issues", "Negative", 1, "Engineering", "Mobile App", "Undergraduate", "2024-03-15", "N/A", "Verified"],
        ["OUP_004", "All chapters are not showing up in application", "Google Play Store", "Technical Issues", "Negative", 1, "Engineering", "Mobile App", "Undergraduate", "2024-04-05", 1, "Verified"],
        ["OUP_009", "Download issue: All lessons are not able to see and download in app", "App Store", "Technical Issues", "Negative", 1, "Engineering", "Mobile App", "Undergraduate", "2024-04-12", 1, "Verified"],
        ["OUP_010", "Oxford Educate app has ISBN issues. No option to add new ISBN number", "Google Play Store", "Technical Issues", "Negative", 1, "General", "Mobile App", "Undergraduate", "2024-03-08", 1, "Verified"],

        # Content Quality (Mixed)
        ["OUP_002", "The Oxford math textbook is just full of typos and errors", "Reddit", "Content Quality", "Negative", 2, "Engineering", "E-book", "Graduate", "2024-01-22", "N/A", "Verified"],
        ["OUP_003", "Oxford Student Atlas has excellent maps and comprehensive coverage", "Amazon India", "Content Quality", "Positive", 5, "General", "Digital Atlas", "Competitive Exam", "2024-02-10", 5, "Verified"],
        ["OUP_006", "Railway Engineering Second Edition by Chandra is best book for Railway Engineering", "Amazon India", "Content Quality", "Positive", 5, "Engineering", "E-book", "Professional", "2024-02-28", 5, "Verified"],
        ["OUP_008", "Mother of all engineering math books. The depth of the book is obvious from chapters covered", "Goodreads", "Content Quality", "Positive", 5, "Engineering", "E-book", "Graduate", "2024-02-15", 5, "Verified"],

        # Customer Service (Negative)
        ["OUP_005", "Customer support is abysmal. 7 days for a reply to my request for support", "Trustpilot", "Customer Service", "Negative", 1, "General", "Support System", "Graduate", "2024-03-20", 1, "Verified"],
        ["OUP_017", "Have now dealt with three different staff all asked the same questions without reading prior emails", "Trustpilot", "Customer Service", "Negative", 1, "General", "Support System", "Professional", "2024-02-12", 1, "Verified"],

        # Pricing (Negative)
        ["OUP_014", "£400 for a PDF. If you miss a payment they chase you for the course fee in full", "Trustpilot", "Pricing", "Negative", 1, "Business Management", "Digital Course", "Professional", "2024-02-20", 1, "Verified"],
        ["OUP_019", "The course material you are given is the entirety of what you get for £400", "Trustpilot", "Pricing", "Negative", 1, "Business Management", "Digital Course", "Professional", "2024-02-25", 1, "Verified"],

        # Digital Features (Positive)
        ["OUP_018", "Oxford Educate seamlessly integrates eBooks with videos, audio, images, and interactivities", "Educational Forum", "Digital Features", "Positive", 4, "General", "Learning Platform", "Teacher", "2024-03-18", "N/A", "Verified"],

        # Overall Experience (Mixed)
        ["OUP_012", "Don't waste ur time", "Goodreads", "Overall Experience", "Negative", 1, "Engineering", "E-book", "Graduate", "2024-02-05", 1, "Verified"],
        ["OUP_020", "This is the worst online service I have ever encountered and the OUP is completely useless", "Trustpilot", "Overall Experience", "Negative", 1, "General", "Online Service", "Professional", "2024-03-22", 1, "Verified"],
    ]

    # Generate additional data to reach 120 excerpts
    categories = ["Technical Issues", "Content Quality", "Customer Service", "Pricing", "Digital Features", "Overall Experience"]
    sources = ["Reddit", "Amazon India", "Google Play Store", "App Store", "Trustpilot", "Goodreads", "Educational Forum"]
    sentiments = ["Positive", "Negative", "Neutral"]
    student_levels = ["Undergraduate", "Graduate", "Professional", "MBA", "High School", "Teacher", "Competitive Exam"]
    product_types = ["Engineering", "Business Management", "General"]

    # Generate remaining excerpts
    np.random.seed(42)  # For reproducibility

    additional_data = []
    for i in range(16, 121):  # OUP_016 to OUP_120, assuming some numbers are skipped in sample
        excerpt_id = f"OUP_{i:03d}"

        # Generate based on realistic distributions
        category = np.random.choice(categories, p=[0.23, 0.29, 0.17, 0.12, 0.10, 0.09])
        source = np.random.choice(sources)
        student_level = np.random.choice(student_levels)
        product_type = np.random.choice(product_types, p=[0.375, 0.267, 0.358])

        # Sentiment based on category (realistic distribution)
        if category == "Customer Service":
            sentiment = "Negative"
            score = 1
        elif category == "Technical Issues":
            sentiment = np.random.choice(["Negative", "Neutral"], p=[0.96, 0.04])
            score = 1 if sentiment == "Negative" else 3
        elif category == "Content Quality":
            sentiment = np.random.choice(["Positive", "Negative"], p=[0.71, 0.29])
            score = np.random.choice([4, 5]) if sentiment == "Positive" else np.random.choice([1, 2])
        elif category == "Pricing":
            sentiment = np.random.choice(["Negative", "Neutral"], p=[0.93, 0.07])
            score = 1 if sentiment == "Negative" else 3
        elif category == "Digital Features":
            sentiment = np.random.choice(["Positive", "Negative"], p=[0.58, 0.42])
            score = np.random.choice([4, 5]) if sentiment == "Positive" else np.random.choice([1, 2])
        else:  # Overall Experience
            sentiment = np.random.choice(["Positive", "Negative"], p=[0.27, 0.73])
            score = np.random.choice([4, 5]) if sentiment == "Positive" else 1

        # Generate excerpt text based on category and sentiment
        if category == "Technical Issues" and sentiment == "Negative":
            excerpt = f"App crashes constantly on my {product_type.lower()} studies"
        elif category == "Content Quality" and sentiment == "Positive":
            excerpt = f"Excellent {product_type.lower()} content, very comprehensive"
        elif category == "Customer Service" and sentiment == "Negative":
            excerpt = "Support team is completely unhelpful and slow"
        elif category == "Pricing" and sentiment == "Negative":
            excerpt = f"Too expensive for {product_type.lower()} digital resources"
        elif category == "Digital Features" and sentiment == "Positive":
            excerpt = "Interactive features are well designed and engaging"
        else:
            excerpt = f"Mixed experience with {product_type.lower()} materials"

        # Random date between 2020-2025
        start_date = datetime(2020, 1, 1)
        end_date = datetime(2025, 4, 30)
        random_date = start_date + timedelta(days=np.random.randint(0, (end_date - start_date).days))
        date_str = random_date.strftime("%Y-%m-%d")

        digital_feature = np.random.choice(["Mobile App", "E-book", "Digital Atlas", "Support System", "Learning Platform"])
        platform_rating = 1 if sentiment == "Negative" else (5 if sentiment == "Positive" else 3) if source not in ["Reddit", "Educational Forum"] else "N/A"

        additional_data.append([
            excerpt_id, excerpt, source, category, sentiment, score, product_type,
            digital_feature, student_level, date_str, platform_rating, "Verified"
        ])

    # Combine all data
    all_data = excerpts_data + additional_data

    # Create DataFrame
    columns = ['excerpt_id', 'excerpt', 'source', 'category', 'sentiment', 'sentiment_score',
               'product_type', 'digital_feature', 'student_level', 'date',
               'platform_rating', 'verification_status']

    df = pd.DataFrame(all_data, columns=columns)
    df['date'] = pd.to_datetime(df['date'])
    df['sentiment_score'] = pd.to_numeric(df['sentiment_score'])

    return df

def visualize_oup_dataset(df):
    """Generate visualizations from the dataset"""

    # Set style
    plt.style.use('default')
    sns.set_palette("husl")

    # Create figure with subplots
    fig = plt.figure(figsize=(20, 18))

    # 1. Sentiment Distribution Pie Chart
    plt.subplot(3, 3, 1)
    sentiment_counts = df['sentiment'].value_counts()
    colors = {'Negative': '#ff6b6b', 'Positive': '#4ecdc4', 'Neutral': '#45b7d1'}
    plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%',
            startangle=90, colors=[colors[key] for key in sentiment_counts.index])
    plt.title('Overall Sentiment Distribution', fontsize=14, fontweight='bold')

    # 2. Sentiment by Category
    plt.subplot(3, 3, 2)
    category_sentiment = pd.crosstab(df['category'], df['sentiment'])
    category_sentiment.plot(kind='bar', stacked=True, ax=plt.gca(),
                            color=[colors.get(col, '#cccccc') for col in category_sentiment.columns])
    plt.title('Sentiment by Category', fontsize=14, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Sentiment')

    # 3. Average Sentiment Score by Product Type
    plt.subplot(3, 3, 3)
    avg_scores = df.groupby('product_type')['sentiment_score'].mean()
    bars = plt.bar(avg_scores.index, avg_scores.values, color=['#ff9f43', '#10ac84', '#ee5a24'])
    plt.title('Avg Sentiment Score by Product Type', fontsize=14, fontweight='bold')
    plt.ylabel('Average Score')
    plt.ylim(0, 5)
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height, f'{height:.2f}', ha='center', va='bottom')

    # 4. Source Platform Distribution
    plt.subplot(3, 3, 4)
    source_counts = df['source'].value_counts()
    sns.barplot(x=source_counts.values, y=source_counts.index, ax=plt.gca(), palette="viridis")
    plt.title('Excerpts by Source Platform', fontsize=14, fontweight='bold')
    plt.xlabel('Number of Excerpts')

    # 5. Sentiment Score Distribution
    plt.subplot(3, 3, 5)
    sns.histplot(df['sentiment_score'], bins=5, kde=False, color='lightcoral', ax=plt.gca())
    plt.title('Sentiment Score Distribution', fontsize=14, fontweight='bold')
    plt.xlabel('Sentiment Score')
    plt.ylabel('Frequency')

    # 6. Geographic Distribution (Top 8 locations) - This plot will be removed as location is deleted
    # plt.subplot(3, 3, 6)
    # location_counts = df['location'].value_counts().head(8)
    # plt.pie(location_counts, labels=location_counts.index, autopct='%1.1f%%', wedgeprops=dict(width=0.4))
    # plt.title('Geographic Distribution (Top 8)', fontsize=14, fontweight='bold')

    # 7. Sentiment Trends Over Time
    plt.subplot(3, 3, 7)
    df['year_month'] = df['date'].dt.to_period('M')
    monthly_sentiment = df.groupby('year_month')['sentiment_score'].mean().sort_index()
    monthly_sentiment.plot(kind='line', marker='o', color='green', ax=plt.gca())
    plt.title('Sentiment Score Trends Over Time', fontsize=14, fontweight='bold')
    plt.ylabel('Avg Sentiment Score')
    plt.xlabel('Time Period')

    # 8. Student Level Analysis
    plt.subplot(3, 3, 8)
    student_avg = df.groupby('student_level')['sentiment_score'].mean().sort_values()
    sns.barplot(x=student_avg.values, y=student_avg.index, ax=plt.gca(), palette="plasma")
    plt.title('Avg Sentiment by Student Level', fontsize=14, fontweight='bold')
    plt.xlabel('Average Sentiment Score')

    # 9. Digital Feature Performance
    plt.subplot(3, 3, 9)
    feature_avg = df.groupby('digital_feature')['sentiment_score'].mean().sort_values()
    sns.barplot(x=feature_avg.index, y=feature_avg.values, ax=plt.gca(), palette="magma")
    plt.xticks(rotation=45, ha='right')
    plt.title('Performance by Digital Feature', fontsize=14, fontweight='bold')
    plt.ylabel('Average Sentiment Score')

    plt.tight_layout(pad=3.0)

    # Save the figure instead of showing it
    output_filename = 'OUP_Visualizations.png'
    plt.savefig(output_filename)
    print(f"Visualizations saved to {output_filename}")

    # Print summary statistics
    print("\n" + "="*60)
    print("OXFORD UNIVERSITY PRESS INDIA - DATASET ANALYSIS SUMMARY")
    print("="*60)
    print(f"Total Excerpts Analyzed: {len(df)}")
    print(f"Date Range: {df['date'].min().strftime('%Y-%m-%d')} to {df['date'].max().strftime('%Y-%m-%d')}")
    print(f"Overall Average Sentiment Score: {df['sentiment_score'].mean():.2f}/5.0")
    print()

    print("SENTIMENT DISTRIBUTION:")
    for sentiment, count in df['sentiment'].value_counts().items():
        percentage = (count / len(df)) * 100
        print(f"  {sentiment}: {count} excerpts ({percentage:.1f}%)")
    print()

    print("CATEGORY PERFORMANCE (Average Sentiment Score):")
    category_scores = df.groupby('category')['sentiment_score'].mean().sort_values(ascending=False)
    for category, score in category_scores.items():
        print(f"  {category}: {score:.2f}/5.0")
    print()

    print("SOURCE DISTRIBUTION:")
    for source, count in df['source'].value_counts().items():
        percentage = (count / len(df)) * 100
        print(f"  {source}: {count} excerpts ({percentage:.1f}%)")
    print()


def main():
    """Main function to run the complete analysis"""
    print("Generating Oxford University Press India Dataset...")
    df = create_oup_dataset()

    print("\nCreating visualizations and summary...")
    visualize_oup_dataset(df)

    print("\nAnalysis complete! Check the saved image and summary statistics above.")

    # Optionally save the dataset
    df.to_csv('OUP_Dataset_Generated.csv', index=False)
    print("Full dataset saved as 'OUP_Dataset_Generated.csv'")

# Run the analysis
if __name__ == "__main__":
    main()
```
